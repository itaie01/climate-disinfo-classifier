```python
# import libraries
import matplotlib.pyplot as plt
from datasets import load_dataset
from collections import Counter
```

```python
dataset = load_dataset("quotaclimat/frugalaichallenge-text-train")
print(next(iter(dataset["train"])))
```

```python
all_train_data = dataset["train"].to_list()
all_train_data[0]
```

```python
all_train_quotes = [datum["quote"] for datum in all_train_data]
all_train_words = " ".join(all_train_quotes).split()
word_freq = Counter(all_train_words)
common_words = word_freq.most_common(20)
```

```python
words = [word[0] for word in common_words]
counts = [word[1] for word in common_words]
```

```python
plt.figure(figsize=(15, 8))
bars = plt.bar(words, counts, color="steelblue", edgecolor="black")
plt.title("Top 20 Most Common Words in Quota Climate Train")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# label distributions
all_train_labels = [datum["label"] for datum in all_train_data]
label_freq = Counter(all_train_labels)
```

```python
plt.figure(figsize=(15, 8))
bars = plt.bar(
    label_freq.keys(), label_freq.values(), color="steelblue", edgecolor="black"
)
plt.title("Label Distribution in Quota Climate Train")
plt.xlabel("Labels")
plt.ylabel("Frequency")

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 0.5,
        int(yval),
        ha="center",
        va="bottom",
        fontsize=12,
    )

plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()

plt.show()
```
